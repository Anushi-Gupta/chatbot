```python
import numpy as np
import nltk             # bundle of libraries for data preprocessig
import random           # generate response random from bot
import string           # removal of punctuation
import bs4 as bs        # convert html ti lxml
import urllib.request   # to scrap the web page
import re               # remove regular expression
```

# Scraping data from tennis wiki website

```python
raw_html = urllib.request.urlopen("https://en.wikipedia.org/wiki/Tennis")
raw_html = raw_html.read()
```

# Raw data looks like:

```python
raw_html
```

# Converting to HTML

```python
article_html = bs.BeautifulSoup(raw_html,"lxml")
```

```python
article_html
```

# Sorting HTML file with the paragraph P

```python
article_paragraphs = article_html.find_all("p")
```

```python
article_paragraphs
```

```python
article_text = ""
```

```python
for para in article_paragraphs:
    article_text +=para.text
```

```python
article_text
```

# Converting all aphabets into smaller case

```python
article_text = article_text.lower()
```

# Removing numbers and symbols

```python
article_text = re.sub(r'\[[0-9]*\]',"",article_text)
#article_text = re.sub(r'\s+','',article_text)
```

```python
article_text
```

# Sentence tokenization

```python
article_sentences = nltk.sent_tokenize(article_text)
article_words = nltk.word_tokenize(article_text)
```

```python
article_sentences
```

```python
article_words
```

# Lemmatization

```python
wnlemmatizer = nltk.stem.WordNetLemmatizer()
```

```python
def perform_lemmatization(tokens):
    return [wnlemmatizer.lemmatize(token) for token in tokens]
    
```

# Removing Punctuation

```python
punctuation_removal = dict((ord(punctuation), None) for punctuation in string.punctuation)

def get_processed_text(document):
    return perform_lemmatization(nltk.word_tokenize(document.lower().translate(punctuation_removal)))
```

```python
greeting_inputs = ("hey","good morning","good evening","morning","evening","hi","hii","whtsup","hello")
greeting_responses = ["hi","good morning","hello","hola"]
```

```python
def generate_greeting_response(greeting):
    for token in greeting.split():
        if token.lower() in greeting_inputs:
            return random.choice(greeting_responses)
```

# Term and inverse document frequency vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def generate_response(user_input):
    datatrainedbot_response = ''
    article_sentences.append(user_input)
    
    word_vectorizer = TfidfVectorizer(tokenizer=get_processed_text, stop_words='english')
    all_word_vectors = word_vectorizer.fit_transform(article_sentences)
    similar_vector_values = cosine_similarity(all_word_vectors[-1], all_word_vectors)
    similar_sentence_number = similar_vector_values.argsort()[0][-2]
    
    matched_vector = similar_vector_values.flatten()
    matched_vector.sort()
    vector_matched = matched_vector[-2]
    
    if vector_matched == 0:
        datatrainedbot_response = datatrainedbot_response + "i am sorry, i could not understsnd you"
        return datatrainedbot_response
    else:
        datatrainedbot_response = datatrainedbot_response + article_sentences[similar_sentence_number]
        return datatrainedbot_response
        
```
